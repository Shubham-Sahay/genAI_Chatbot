```python
import pandas as pd
```

```python
# Loading the train and test datasets (With predictions)
predicted_train = pd.read_csv("datasets/predicted_train.csv")
predicted_test = pd.read_csv("datasets/predicted_test.csv")
```

# Processing Train and Test Data
For train data we can delete the following columns
1. primary_close_flag
2. final_close_flag

These columns would be basically replaced by their predicted versions.

For test data, we would keep all the columns

```python
predicted_train = (predicted_train
                   .drop(["primary_close_flag", "final_close_flag"], axis=1)
                   .rename({"pred_primary_close_flag": "primary_close_flag",
                            "pred_final_close_flag": "final_close_flag"},
                           axis=1
                          )
                  )

predicted_train.info()
```

```python
predicted_test.info()
```

# Concatenating Train and Test

```python
final_df = pd.concat([predicted_train, predicted_test])
final_df.info()
```

```python
# Saving as CSV
final_df.to_csv("datasets/loan_application_data_combined.csv",index=False)`
```

```python
import chainlit as cl
from sql import sql_query_engine
from llm import llm, embed_model
from llama_index.callbacks import CallbackManager, LlamaDebugHandler

from llama_index.tools import QueryEngineTool, ToolMetadata
from llama_index.query_engine import SubQuestionQueryEngine
from llama_index import set_global_service_context, ServiceContext


# llama_debug = LlamaDebugHandler(print_trace_on_end=True)
# callback_manager = CallbackManager([llama_debug])
# service_context = ServiceContext.from_defaults(callback_manager=callback_manager, 
                                        #        llm=llm, embed_model=embed_model)

service_context = ServiceContext.from_defaults(llm=llm, embed_model=embed_model)

set_global_service_context(service_context)

query_engine_tools = [
        QueryEngineTool(
                query_engine=sql_query_engine,
                metadata = ToolMetadata(
                        name="Customer Loan Database",
                        description= '''
                        Database containing the Loan Application Data table
                        '''
                )
        )
]

subquery_engine = SubQuestionQueryEngine.from_defaults(query_engine_tools=query_engine_tools,
                                                       service_context = service_context,
                                                       use_async = True, verbose=True
                                                       )

response = subquery_engine.query('what is the average and total loan amount. Also tell me about the unique number of applicants.')
```
